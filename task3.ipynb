# Стьюдент и Манна-Уитни 

 $X_i \sim \mathcal{N}(0, 1)$, $Y_i \sim \mathcal{N}(0, 1)$, $n = 50$

```python
X = []
Y = []
for i in range(100):
    X.append(stats.norm(0, 1).rvs(50))
    Y.append(stats.norm(0, 1).rvs(50))
```

```python
pv_tt = [stats.ttest_ind(x, y)[1] for x,y in zip(X, Y)]
pv_tt.sort()
pv_mw = [stats.mannwhitneyu(x, y, alternative='two-sided')[1] for x,y in zip(X, Y)]
pv_mw.sort()
plt.plot(pv_tt, [i / 99 for i in range(100)], label = 't-test')
plt.plot(pv_mw, [i / 99 for i in range(100)], label = 'M-W')
plt.legend()
plt.show()
```

# ANOVA и Краскелла-Уоллиса


$X_i \sim \mathcal{N}(0, 1)$, $Y_i \sim \mathcal{N}(0, 1)$, $Z_i \sim \mathcal{N}(0, 1)$, $n = 50$

```python
X = []
Y = []
Z = []
for i in range(100):
    X.append(stats.norm(0, 1).rvs(size = 50))
    Y.append(stats.norm(0, 1).rvs(size = 50))
    Z.append(stats.norm(0, 1).rvs(size = 50))
pv_anova = [stats.f_oneway(x, y, z)[1] for x, y, z in zip(X, Y, Z)]
pv_anova.sort()
pv_kw = [stats.kruskal(x, y, z)[1] for x, y, z in zip(X, Y, Z)]
pv_kw.sort()
plt.plot(pv_anova, [i / 99 for i in range(100)], label = 'ANOVA')
plt.plot(pv_kw, [i / 99 for i in range(100)], label = 'K-W')
plt.legend()
plt.show()
```
